We start with a proporitional contoller. $C(s) = K$

```python
import control
import matplotlib.pyplot as plt
from IPython.html.widgets import *

%matplotlib inline
num = [1]
den = [1,8,12,0]
P = control.tf(num,den)

def proportional_controller(K):
    C = control.tf([K],[1])
    control.bode(P*C,dB = True )
    plt.show()
interact(proportional_controller,K=(1,300))
```

# Lead controller

```python
def lead_controller(a,b):
    C = control.tf([1,-10**a],[1,-10**b])
    control.bode(C,dB = True )
    plt.show()
interact(lead_controller,a=(1,3,1),b=(3,6,1))
    
```

```python
def lag_conroller(a,b):
    C = control.tf([1,-10**a],[1,-10**b])
    control.bode(C,dB = True )
    plt.show()
interact(lead_controller,a=(3,6,1),b=(1,3,1))
```
