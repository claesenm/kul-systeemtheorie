```python
%matplotlib inline
from control import*
from matplotlib import*
from pylab import*
from math import*
from numpy import*
from IPython.html.widgets import*


    
  
```

```python
def first_order_system(k,tau):
    """Plots the poles and zeros of a first order system"""
    TF=TransferFunction([k,],[tau,1])
    print "H(s) ="
    print TF
    zeros = zero(TF)
    poles = pole(TF)
    subplot(1,2,1)
    matlab.pzmap(TF)
    
    subplot(1,2,2)
    t,y=step_response(TF)
    plot(t,y,label='Step Response')
    title('Step Response')
    xlabel('Time(sec)')
    ylabel('Magnitude')
    
    fig= gcf()
    fig.set_size_inches(15,5)
    show ()
    
  
    bode(TF,label='Bode Plot') 
    title('Bode Plot')
    fig= gcf()
    fig.set_size_inches(9,9)
    show ()
    
def second_order_system(zeta,natfreq):
    """Plots the poles and zeros of a second order system"""
    TF=TransferFunction([natfreq**2,],[1,2*zeta*natfreq,natfreq**2])
    print "H(s) ="
    print TF
    zeros = zero(TF)
    poles = pole(TF)
    subplot(1,2,1)
    matlab.pzmap(TF)
   
    subplot(1,2,2)
    t,y=step_response(TF)
    plot(t,y,label = 'Step Response')
    title('Step Response')
    xlabel('Time(sec)')
    ylabel('Magnitude')
    
    fig= gcf()
    fig.set_size_inches(15,5)
    show()
    
    bode(TF,label='Bode Plot')
    title('Bode Plot')
    fig= gcf()
    fig.set_size_inches(9,9)
    show ()


    
   
    

```

#Effect of the time constant, natural frequency and damping ratio demo
In this demo you can see what happens what the influence is of the time constant on first order systems
<br>
<br> $= \large{\frac{k}{\tau s+1}}$ (with $\tau$ the symbol for time constant)
<br>
<br>and what the effect of the natural frequency and damping ratio is on second order systems
<br>
<br>$=\large{\frac{\omega_n ^2}{s^2 +2\zeta\omega_n+\omega_n ^2}}$ (with $\omega_n$ the natural frequency and
 $\zeta$ the damping ratio).
<br>
<br>There are three possible plots: Zeros/poles plot, Bode plot and step response.

<br>A first order system always has $1$ real pole. 
<br>
<br>The bigger $\tau$ gets, the closer the pole gets to 0.
<br>
<br>A standard first order system has 1 breaking point, $\frac{1}{\tau}$. Thus changing $\tau$ changes the
<br>
<br>magnitude plot and the phase plot of the Bode plot.
<br>
<br>For a positive $\tau$, the step response converges to a certain value.
