# Sampling

## Signal

```python
from sampling import *
%matplotlib inline
N = 500
T = 1/700.0
t = np.linspace(-N*T/2.0,N*T/2.0,N)
y = np.sin(2*np.pi*100*t)/t
# t,y = input_selector()
plt.plot(t,y)
plt.show()
```

## Fourier Transform

```python
yf = 2.0/N * np.abs(fourier.fft(y))
print len(yf)
xf = np.fft.fftfreq(len(y), t[1] - t[0])
plt.plot(xf,yf)
plt.show()
```
