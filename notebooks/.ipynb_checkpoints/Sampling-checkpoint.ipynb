# Sampling

```python
%matplotlib inline
from math import *
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fourier

x = np.linspace(-10,10,1000)
signal = [0,]*450 + [1,]*100 + [0,]*450
plt.plot(x,signal)
plt.show()
y = fourier.rfft(signal)
signal_2 = y[0]*x+ y[1]*np.sin(x)+y[2]*np.cos(x)+y[3]*np.sin(2*x)
plt.plot(x,fourier.fftshift(y))
plt.xlim(-3,3)
plt.show()
```
